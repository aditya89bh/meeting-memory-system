# 01 — Importing meetings

Turn raw transcripts into structured, queryable memory. This notebook imports a bundled example organization into a throwaway database and inspects what was captured.

```python
import pathlib
import tempfile

from meeting_memory.sdk import MeetingMemoryClient

REPO = pathlib.Path.cwd().parent if pathlib.Path.cwd().name == "notebooks" else pathlib.Path.cwd()
DB = pathlib.Path(tempfile.mkdtemp()) / "memory.db"
DATA = REPO / "examples" / "organizations" / "saas"
print("database:", DB)
```

```python
client = MeetingMemoryClient.local(DB)
result = client.import_directory(str(DATA), recursive=True)
result
```

## Inspect stored memory

```python
stats = client.stats()
print("meetings:", stats["meetings"], "memories:", stats["memories"])
stats["by_type"]
```

```python
for meeting in client.meetings(limit=10)["items"]:
    print(meeting["date"], "-", meeting["title"])
```
