# 02 — Searching memory

Run ranked retrieval over imported memory and inspect why results matched.

```python
import pathlib
import tempfile

from meeting_memory.sdk import MeetingMemoryClient

REPO = pathlib.Path.cwd().parent if pathlib.Path.cwd().name == "notebooks" else pathlib.Path.cwd()
DB = pathlib.Path(tempfile.mkdtemp()) / "memory.db"
client = MeetingMemoryClient.local(DB)
client.import_directory(str(REPO / "examples" / "organizations" / "enterprise"), recursive=True)
print("ready")
```

```python
response = client.search("migrate core services", limit=5)
for hit in response["results"]:
    m = hit["memory"]
    print(f"{hit['score']:.3f}  [{m['memory_type']}] {m['speaker']}: {m['text']}")
```

## Filter by type and explain a match

```python
risks = client.search("risk", limit=5, memory_type=["risk"])
for hit in risks["results"]:
    print(round(hit["score"], 3), hit["memory"]["text"])
```

```python
top = response["results"][0]
print("explanation:", top["explanation"])
print("from meeting:", top["meeting"]["title"])
```
