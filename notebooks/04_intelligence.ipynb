# 04 — Organizational intelligence

Surface recurring risks, revisited decisions, stale commitments, and an overall health score.

```python
import pathlib
import tempfile

from meeting_memory.sdk import MeetingMemoryClient

REPO = pathlib.Path.cwd().parent if pathlib.Path.cwd().name == "notebooks" else pathlib.Path.cwd()
DB = pathlib.Path(tempfile.mkdtemp()) / "memory.db"
client = MeetingMemoryClient.local(DB)
client.import_directory(str(REPO / "examples" / "organizations" / "research-lab"), recursive=True)
print("ready")
```

```python
metrics = client.metrics()
print("overall health:", round(metrics["overall"], 3))
metrics["scores"]
```

## Discovered insights

```python
for insight in client.insights()["items"]:
    print(f"[{insight['severity']}] {insight['title']}")
```

## Prioritized recommendations

```python
for rec in client.recommendations()["items"]:
    print("-", rec["title"])
```

## The full report

```python
report = client.report(fmt="markdown")
print(report["content"][:600])
```
