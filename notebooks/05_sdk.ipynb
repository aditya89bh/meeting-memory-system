# 05 — The Python SDK

`MeetingMemoryClient` has one method per capability and two interchangeable transports: in-process **local** mode and **HTTP** mode. This notebook uses local mode, which needs no running server.

````python
import pathlib, tempfile
from meeting_memory.sdk import MeetingMemoryClient

REPO = pathlib.Path.cwd().parent if pathlib.Path.cwd().name == "notebooks" else pathlib.Path.cwd()
DB = pathlib.Path(tempfile.mkdtemp()) / "memory.db"
client = MeetingMemoryClient.local(DB)
print("mode:", client.mode)
print("health:", client.health())
print("version:", client.version())
````

## A complete workflow in one place

````python
client.import_directory(str(REPO / "examples" / "organizations" / "university"), recursive=True)
stats = client.stats()
print("meetings:", stats["meetings"], "memories:", stats["memories"])

for hit in client.search("accreditation", limit=3)["results"]:
    print(round(hit["score"], 3), hit["memory"]["text"])

g = client.graph()
print("graph:", g["nodes"], "nodes /", g["edges"], "edges")

print(client.report(fmt="markdown")["content"][:200])
````

## Switching to HTTP mode

The same code works against a running server — only the constructor changes:

```python
with MeetingMemoryClient.connect("http://127.0.0.1:8000") as client:
    client.search("accreditation", limit=3)
```

Start a server first with `python examples/api/serve.py --db atlas.db --port 8000`.
