# 03 — The organizational graph

Explore people, projects, meetings, decisions, and risks as a connected graph.

```python
import pathlib
import tempfile

from meeting_memory.sdk import MeetingMemoryClient

REPO = pathlib.Path.cwd().parent if pathlib.Path.cwd().name == "notebooks" else pathlib.Path.cwd()
DB = pathlib.Path(tempfile.mkdtemp()) / "memory.db"
client = MeetingMemoryClient.local(DB)
client.import_directory(str(REPO / "examples" / "organizations" / "saas"), recursive=True)
print("ready")
```

```python
summary = client.graph(limit=5)
print("nodes:", summary["nodes"], "edges:", summary["edges"])
print("by node type:", summary["by_node_type"])
print("by relationship:", summary["by_relationship"])
```

## Explore a person's neighbourhood

```python
people = client.graph(node_type=["person"], limit=10)["listed"]
for node in people:
    print(node["node_id"], "->", node["label"])
```

```python
node_id = people[0]["node_id"]
neighbourhood = client.neighbors(node_id, depth=1)
neighbourhood.get("by_relationship", neighbourhood)
```

## Export the graph

```python
full = client.graph()
print("exported", full["nodes"], "nodes and", full["edges"], "edges")
```
